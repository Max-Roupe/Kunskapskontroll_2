### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OO

x = np.linspace(-4, 4, 100)

y = x**2

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, y, label='$y = x^2$', color="green")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Object-Orianted")
ax.legend()
```

```python
# pyplot

x = np.linspace(-4 , 4, 1000)

y = x**2

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, y, label='$y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Pyplot approach")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axd = plt.subplot_mosaic([['left', 'right']], layout='constrained', figsize=(12, 6))
axd['left']
axd['right']
plt.suptitle("Scatter Plot & Bar Plot", fontsize=18)

axd["right"].bar(names, values, color="red", label=r"$Fruits sold$")
axd["right"].set_title("Bar Plot", loc="center", weight="bold", fontsize=15)
axd["right"].set_xlabel("Fruits", weight="heavy", fontsize=12)
axd["right"].set_ylabel("Quantity", rotation=0, weight="heavy", fontsize=9)
axd["right"].legend(loc=0)

axd["left"].scatter(random_data_x, random_data_y, color="green", alpha=0.7)
axd["left"].plot(x, y, label=r"$y = x^2$", color=("orange"))
axd["left"].set_title("Scatter Plot", weight="bold", fontsize=15)
axd["left"].set_xlabel("X", weight="bold", fontsize=12)
axd["left"].set_ylabel("Y", rotation = 0, weight="heavy", fontsize=12)
axd["left"].legend(loc=3)
```
